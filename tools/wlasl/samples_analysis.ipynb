### Imports

```python
import json
import os
import shutil
import pandas as pd
import matplotlib.pyplot as plt
```

## WLASL100

```python
with open('nslt_100.json') as fin:
    videos = json.load(fin)
```

```python
videos
```

```python
df = pd.DataFrame(videos).transpose()
df['class'] = df.apply(lambda row: row.action[0], axis=1)
df.head()
```

```python
df[df['subset']=='train']['class'].value_counts().mean()
```

```python
df[df['subset']=='train']['class'].value_counts().min()
```

```python
df[df['subset']=='train']['class'].value_counts()
```

```python
plt.plot(df[df['subset']=='train']['class'].value_counts())
plt.xlabel('class_id')
plt.ylabel('number of samples')
plt.title('train samples vs classes')
```

### Train + Validation

```python
df[(df['subset']=='train') | (df['subset']=='val')]['class'].value_counts()
```

```python
df[(df['subset']=='train') | (df['subset']=='val')]['class'].value_counts().min()
```

```python
df[(df['subset']=='train') | (df['subset']=='val')]['class'].value_counts().mean()
```

```python
plt.plot(df[(df['subset']=='train') | (df['subset']=='val')]['class'].value_counts())
plt.xlabel('class_id')
plt.ylabel('number of samples')
plt.title('train+val samples vs classes')
```

## WLASL2000

```python
with open('nslt_100.json') as fin:
    videos = json.load(fin)
    
df = pd.DataFrame(videos).transpose()
df['class'] = df.apply(lambda row: row.action[0], axis=1)

plt.plot(df['class'].value_counts())
plt.xlabel('class_id')
plt.ylabel('number of samples')
plt.title('samples vs classes')
```
