```python
import json
import os
import shutil
```

```python
DATA_PATH = "../../data/wlasl/wlasl-uncompressed"
os.chdir(DATA_PATH)
```

```python
with open('wlasl_2000.json') as fin:
    videos = json.load(fin)
```

```python
videos
```

```python
os.makedirs('train', exist_ok=True)
os.makedirs('test', exist_ok=True)
os.makedirs('val', exist_ok=True)
```

```python
for video_id in videos:
    subset = videos[video_id]['subset']
    try:
        os.rename(f'videos/{video_id}.mp4', f'{subset}/{video_id}.mp4')
    except FileNotFoundError:
        print(f'{video_id}.mp4 not found. Skipping and going next...')
```
